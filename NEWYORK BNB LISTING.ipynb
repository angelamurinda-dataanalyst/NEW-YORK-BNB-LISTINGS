```python
#finding correlation between some numerical columns using correlation heatmap chart
```

```python
bnb_listing.dtypes
```

```python
corr=bnb_listing[['latitude','longitude','price','minimum_nights','number_of_reviews','beds','availability_365']].corr()
```

```python
corr
```

```python
plt.figure(figsize=(10,7))
```

```python
sns.heatmap(data=corr, annot=True)
```

```python
plt.title('CORRELATION BETWEEN COLUMNS')
```

```python
plt.show()
```

```python
#from the heat map strong colors like red nad white how strong correlation while purple to darker shade show little to no correlation
#price and beds have a moderate positive correlation; as the price increases number of bed increases
#price and longitude have a weak negative correlation; price tends to decrease as longitude increases
#other variables like number of reviews and minimum nights have little to no correlation with price,
#Therefore, the number of reviews and availability have no impact on the price
```
